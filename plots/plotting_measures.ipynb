```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from decouple import config
import seaborn as sns
sns.set(font_scale=1)
```

```python
data = config('data')
os.chdir(data) 
```

```python
plotting_df = pd.read_csv('cortical_measures.csv')

groups = plotting_df.groupby('age_adjusted_group')
aan = groups.get_group('AAN')
wr = groups.get_group('WR')
hc = groups.get_group('HC') 
```

```python
fig,ax = plt.subplots(1,3, figsize=(16,8))
fig.suptitle('Group Differences in Global Measures')

sns.boxplot(ax = ax[0], x = "age_adjusted_group", y = "TotalGrayVol", data = plotting_df, linewidth=1, 
            showfliers=False)

sns.stripplot(ax = ax[0], x = "age_adjusted_group", y = "TotalGrayVol", data=plotting_df, color=".15")
ax[0].set(xlabel='Groups', ylabel='Gray matter volume (mm3)')


sns.boxplot(ax=ax[1], x="age_adjusted_group", y="mean_lgi", data=plotting_df, linewidth=1, 
            showfliers=False)

sns.stripplot(ax=ax[1], x="age_adjusted_group", y="mean_lgi", data=plotting_df, color=".15")
ax[1].set(xlabel='Groups', ylabel='LGI (arbitrary units)')

sns.boxplot(ax=ax[2], x="age_adjusted_group", y="mean_thickness", data=plotting_df, linewidth=1, 
            showfliers=False)
sns.stripplot(ax=ax[2], x="age_adjusted_group", y="mean_thickness", data=plotting_df, color=".15")
ax[2].set(xlabel='Groups', ylabel='Thickness (mm)')
```

```python
fig,ax = plt.subplots(1,2, figsize=(16,8))
fig.suptitle('Group Differences in Global Measures')

sns.boxplot(ax=ax[0], x="age_adjusted_group", y="postcentral", data=plotting_df, linewidth=1, 
            showfliers=False)
sns.stripplot(ax=ax[0], x= "age_adjusted_group", y="postcentral", data=plotting_df, color=".15")
ax[0].set(xlabel = 'Groups', ylabel = 'Parameter Estimates for postcentral gyrus')


sns.boxplot(ax=ax[1], x="age_adjusted_group", y="supramarginal", data=plotting_df, linewidth=1, 
            showfliers=False)
sns.stripplot(ax=ax[1], x="age_adjusted_group", y="supramarginal", data=plotting_df, color=".15")
ax[1].set(xlabel='Groups', ylabel='Parameter Estimates for supramarginal gyrus')
```

```python
fig,ax = plt.subplots(1,1, figsize=(8,8))
plot = sns.regplot(x='Age', y ='mean_lgi', data=aan, 
            scatter_kws={'color':'black'}, 
            line_kws={'color': 'b'},
            ci=False,
           )

plot.set(xlabel='Age for AAN Participants', ylabel='Parameter Estimates for supramarginal gyrus')
```

```python
fig,ax = plt.subplots(1,2, figsize=(16,8))
fig.suptitle('Correlation plots for AAN group and LGI parameter estimates')

sns.regplot(ax=ax[0], x='Age', y='supramarginal', data= aan, 
            scatter_kws={'color':'black'}, 
            line_kws={'color': 'b'},
            ci=False,
           )


sns.regplot(ax=ax[1], x='Age', y='postcentral', data=aan , 
            scatter_kws={'color':'black'}, 
            line_kws={'color': 'b'},
            ci=False,
           )

ax[0].set(xlabel='Age For AAN Group', ylabel='Parameter Estimates for supramarginal gyrus')
ax[1].set(xlabel='Age For AAN Group', ylabel='Parameter Estimates for postcentral gyrus')
```
